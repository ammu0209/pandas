```python

import pandas as pd
#creating pandas series from list
data=[0.25, 0.5, 0.75, 1.0]
series= pd.Series(data,index=['a','b','c','d'],dtype="float",name="python")
print(series)
```

```python
#"Pandas Series from Dictionary
dict={"name":['python','c','c++','java'],"prop":[12,13,14,15]}
series=pd.Series(dict)
series
```

```python
#Arithmetic Operations on Pandas Series
s1=pd.Series(12,index=[1,2,3,4,5,6])
s2=pd.Series(12,index=[1,2,3,4,5])
s1+s2
```

```python
#Filtering Pandas Series Using Boolean Masking".
import pandas as pd
songs3=pd.Series([145 , 142 , 38 , 13],index=['paual','john','george','ringo'])
print(songs3)
mask = songs3 > songs3 . median () # boolean array
print(mask)
songs3 [ mask ]      #pass the mask into an index to access the true values

```

```python
#create a series with the temperature values for the last seven days. Filter out
#the values below the mean.

import pandas as pd
temp=pd.Series([45,54,63,21,34],index=['D1','D2','D3','D4','D5'])
temp
mask=temp<temp.mean()
mask
temp[mask]
```

```python
#create a series with your favorite colors. Use a categorical type
import pandas as pd

# Define favorite colors
colors = pd.Series(["Blue", "Black", "Red", "Blue", "White"], dtype="category")

# Print the Series
print(colors)

# Print category details
print("\nCategories:", colors.cat.categories)
```

```python
#If you want to work with a column without affecting the original DataFrame,
 #you can create an explicit copy of the Series using the copy() method.
# Create a copy of column 'A'
column_a_copy = df['a'].copy()
column_a_copy[0] = 99
print(df)  # Original DataFrame remains unchanged
print(column_a_copy)

```
